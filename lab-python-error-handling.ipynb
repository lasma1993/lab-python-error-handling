````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try: #allows the program to attempt a block of code and catch exceptions (errors) if they occur
                quantity = int(input(f"Enter quantity for {product}: "))#If the input cannot be converted to an integer, a ValueError will be raised, which is handled by the except block.
                if quantity >=0:
                    inventory[product] = quantity
                    valid_input = True #This line sets the valid_input flag to True, which will exit the while loop.
                else: #code that runs if the quantity is negative
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError: #(e.g., if the user enters a non-integer value).
                print("Invalid input. Please enter a valid quantity.")
    return inventory

#Products - outside the function, passed as argument
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products) #storing returned inventory in a variable 
````

````python
def get_customer_orders(inventory):
    customer_orders = set()
    
    # Get the number of orders from the user
    while True:
        try:
            order_input = int(input("Please enter the number of orders: "))
            if order_input > 0:
                break  # Stop the loop
            else:
                print("Quantity cannot be negative. Please enter a valid quantity.")
        except ValueError:
            print("Invalid input. Please enter a valid quantity.")
    
    # Get the specific products for each order
    for i in range(order_input):
        valid_input = False
        while not valid_input:
            product_input = input("Please enter a product: ")
            if product_input in inventory:
                if inventory[product_input] > 0:
                    customer_orders.add(product_input)
                    valid_input = True
                else:
                    print("Product is out of stock, please select a new product.")
            else:
                print("This product does not exist in inventory, please select a new product.")
    
    return customer_orders



customer_orders = get_customer_orders(inventory) 
````
